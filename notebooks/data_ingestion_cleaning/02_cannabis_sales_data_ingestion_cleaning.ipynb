```python
columns_to_keep = [
    "REF_DATE",
    "GEO",
    "DGUID",
    "Type of cannabis",
    "UOM",
    "UOM_ID",
    "SCALAR_FACTOR",
    "SCALAR_ID",
    "VALUE"
]

sales_subset_df = sales_df[columns_to_keep]
```

```python
sales_subset_df
```

```python
sales_subset_df = sales_df[columns_to_keep].copy()
```

```python
sales_subset_df
```

```python
sales_subset_df['Actual_Sales_CAD'] = sales_subset_df['VALUE'] * 1000
```

```python
reported_total_sales = sales_subset_df.loc[sales_subset_df['Type of cannabis'] == 'Total cannabis products', 'Actual_Sales_CAD'].iloc[0]
```

```python
component_sales_df = sales_subset_df[sales_subset_df['Type of cannabis'] != 'Total cannabis products'].copy()
sum_of_component_sales = component_sales_df['Actual_Sales_CAD'].sum()
```

```python
component_sales_df['Sales_Share_%'] = (component_sales_df['Actual_Sales_CAD'] / reported_total_sales) * 100
```

```python
# 5. Print Summary of Calculations
print("\n" + "-"*40)
print("--- Sales Calculation Summary (2021/2022, Canada) ---")
print(f"Reported Total Sales (CAD): {reported_total_sales:,.0f}")
print(f"Sum of Listed Component Sales (CAD): {sum_of_component_sales:,.0f}")
print(f"Difference (CAD): {reported_total_sales - sum_of_component_sales:,.0f}")
print("-" * 40 + "\n")
```
